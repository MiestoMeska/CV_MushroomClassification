### Learning rate 1e-5

```python
batch_size = 16
learning_rate = 1e-5
train_model(ResNetModel, batch_size, learning_rate)
```

```python
batch_size = 32
learning_rate = 1e-5
train_model(ResNetModel, batch_size, learning_rate)
```

```python
batch_size = 64
learning_rate = 1e-5
train_model(ResNetModel, batch_size, learning_rate)
```

```python
batch_size = 64
learning_rate = 1e-5
accumulation_steps = 2
train_model(ResNetModel, batch_size, learning_rate, accumulation_steps )
```

```python
batch_size = 64
learning_rate = 1e-5
accumulation_steps = 3
train_model(ResNetModel, batch_size, learning_rate, accumulation_steps )
```

#### 1e-5 Findings

Observations:

The model with accumulation_steps = 1 achieved an accuracy of approximately 82%, with consistent precision, recall, and F1-score across different classes.

Increasing the accumulation steps to 2 improved the accuracy to around 84%. This indicates that accumulating gradients over multiple batches helps improve model performance.

Further increasing the accumulation steps to 3 led to a slightly higher accuracy of approximately 85%. This suggests that increasing the accumulation steps beyond a certain threshold might not significantly improve performance, as the gains become marginal.

Conclusions:

The use of accumulation steps effectively addresses GPU memory limitations, allowing training with larger effective batch sizes.

Models trained with higher accumulation steps tend to achieve better accuracy, indicating that leveraging accumulated gradients helps improve optimization, especially when GPU memory is constrained.

However, there seems to be diminishing returns with increasing accumulation steps, as the improvements in accuracy become less significant with higher accumulation values.

## Conclusions

### Impact of Batch Size and Learning Rate Without Accumulation Steps


Learning Rate of 1e-3:

Increasing the batch size from 16 to 128 generally improves the model's performance, with accuracy enhancing from 0.74 to 0.87. The improvement is attributed to the model's ability to generalize better with larger batch sizes, reducing overfitting.

Precision, recall, and F1-scores increase with batch size, indicating more balanced performance across classes. The highest jump in performance is observed when batch size increases from 64 to 128.

Learning Rate of 1e-4:

The learning rate of 1e-4 outperforms 1e-3 across all batch sizes, suggesting that a slower learning rate aids in more effective optimization, particularly in finer adjustments in the model's weights.

The model reaches its peak performance at a batch size of 16 with an accuracy of 0.90. However, even larger batch sizes maintain high accuracy levels (around 0.89), showing the robustness of this learning rate setting.

Learning Rate of 1e-5:

The lowest learning rate tested shows a mixed impact. While it generally supports high precision and recall across classes, the accuracy peaks at a batch size of 32 (0.85) and then slightly decreases with larger batch sizes.

This learning rate may require more epochs to converge, especially for larger batch sizes, as indicated by the performance peaks at different epochs for different batch sizes.

### Impact of Gradient Accumulation Steps

Gradient accumulation steps were tested with a batch size of 64 and a learning rate of 1e-5. The use of accumulation steps enhances model performance slightly but noticeably:


**Accumulation Steps = 2:**

The model shows improved performance compared to the configuration without accumulation steps at the same batch size and learning rate, achieving an accuracy of 0.84.

There is a notable balance between precision and recall across different classes, indicating effective learning even with fewer updates.

**Accumulation Steps = 3:**

Further increasing the accumulation steps to 3 improves the model's accuracy to 0.85, with similar trends in precision, recall, and F1-score.
This suggests that more accumulation steps allow for more stable and effective gradient updates, especially beneficial when training with larger batch sizes and lower learning rates.
